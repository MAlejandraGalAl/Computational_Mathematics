```python
#Cálculo del área, perímetro
import math
import numpy as np
import matplotlib.pyplot as plt


def Area_Manual(x,y):
    suma=0
    valorparcial=[]
    for i in range (1,len(x)-1):
        valorparcial.append(y[i]*(x[i+1]-x[i-1]))
    area=(sum(valorparcial)+valorparcial[0])/2 
    return(area)

def Perimetro(x,y):
    distancias=[]
    for i in range(0,len(x)-1):
        distancias.append(math.sqrt((x[i+1]-x[i])**2 + (y[i+1]-y[i])**2))
    perimetro=sum(distancias)
    return(perimetro)

def ordenapuntos (x,y):
    angulos=[]
    xor=[] #x ordenada
    yor=[] #y ordenada
    centroide=(sum(x)/len(x),sum(y)/len(y))
    for i in range(0,len(x)): #foto que adjunto con el código :)
        longCp= math.sqrt((x[i]-centroide[0])**2 + (y[i]-centroide[1])**2)
        puntoAp= (x[i],centroide[1])
        longAp= math.sqrt((centroide[0]-puntoAp[0])**2 + (centroide[1]-puntoAp[1])**2)
        anguloparcial=math.acos((longAp/longCp))
        if x[i]>=centroide[0] and y[i]>=centroide[1]:
            angulo= math.degrees(anguloparcial)
        if x[i]<centroide[0] and y[i]>=centroide[1]:
            angulo= 180-math.degrees(anguloparcial)
        if x[i]<centroide[0] and y[i]<centroide[1]:
            angulo= 180+math.degrees(anguloparcial)
        if x[i]>=centroide[0] and y[i]<centroide[1]:
            angulo= 360-math.degrees(anguloparcial)
        angulos.append(angulo)
    angulord=sorted(angulos,reverse=True) #angulo ordenado
    for i in range(len(angulord)):
        xor.append(x[angulos.index(angulord[i])])
        yor.append(y[angulos.index(angulord[i])])
    xor.append(xor[0])
    yor.append(yor[0])
    return([xor,yor])
```

```python
#Ejemplo facil de calcular
x1=[4,4,2,2] 
y1=[5,3,3,5]
puntosbuenos1=ordenapuntos(x1,y1)
print(puntosbuenos1)
print("Areap= ",Area_Manual(puntosbuenos1[0],puntosbuenos1[1]))
print("Perimetro: ", Perimetro(puntosbuenos1[0],puntosbuenos1[1]))
#grafica
fig, ax = plt.subplots()
ax.plot(puntosbuenos1[0], puntosbuenos1[1], '-o')
plt.show()
```
